```python
# Importando bibliotecas

import pandas as pd
import numpy as np
from scipy.stats import shapiro
import matplotlib.pyplot as plt
```

```python
# Importando os dados

indicadores = pd.read_csv('/content/drive/MyDrive/Prova_datascience/Indicators.csv')
indicadores
```

```python
paises_al = codigos_pais = ["BRA", "MEX", "FRA", "DEU", "JPN", "CHN", "USA", "CAN"]
```

```python
agricultural_growth = indicadores[(indicadores['CountryCode'].isin(paises_al)) & (indicadores['IndicatorName'] == 
'Agriculture, value added (annual % growth)')]
pivot_agricultural_growth = pd.pivot_table(agricultural_growth, values = 'Value', index=['Year'], columns='CountryName')
pivot_agricultural_growth.boxplot(figsize=(20,12))
```

```python
agricultural = indicadores[(indicadores['CountryCode'].isin(paises_al)) & (indicadores['IndicatorName'] == 
'Agricultural machinery, tractors')]
pivot_agricultural = pd.pivot_table(agricultural, values = 'Value', index=['Year'], columns='CountryName')
pivot_agricultural.boxplot(figsize=(20,12))
```

```python
ondedataset =pd.merge(pivot_agricultural,pivot_agricultural_growth, left_index=True, right_index=True )
ondedataset.boxplot(figsize=(30,22))
```

```python
prod_br = ondedataset[['France_x','France_y']]
prod_br.boxplot()
prod_br.plot(figsize=(16,12))
```

```python
prod_br['France_x'].hist()
```

```python
import scipy.stats as st
shapiro_results = st.shapiro(prod_br['France_x'])
shapiro_results
```

## **Resposta 1** ✅

H0: Os dados são normalmente distribuídos Aceitar H0 é p_valor > 0,05

pvalue=1.0 é **maior** que 0.05, a hipótese nula é **aceita**.
